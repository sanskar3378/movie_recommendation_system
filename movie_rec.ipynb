```python
import ast

ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(data):
    lst = []
    for i in ast.literal_eval(data):
        lst.append(i['name'])
    return lst
```

```python
def convert2 (data):
    lst = []
    data = ast.literal_eval(data)
    count = 0
    for i in data:
        if count != 10:
            count += 1
            lst.append(i['name'])
        else:
            break

    return lst  
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head()
```

```python
# Fetching director name from crew
```

```python
def get_director(data):
    lst = []
    data = ast.literal_eval(data)
    for i in data:
        if(i['job'] == 'Director'):
            lst.append(i['name'])
    return lst
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
movies['genre'] = movies['genres'].apply(lambda x : [i.replace(' ', '_') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(' ', '_') for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(' ', '_') for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(' ', '_') for i in x])
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split(' '))
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].str.join(' ')
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

now we have to convert the tags from string to vectors.
for this I am using text vectorisation. For this I am using scikit-learn library's CounterVectorizer method

CountVectorizor - It converts text documents into a numerical matrix based on word frequency, which helps in analyzing large collections of text data. This is done by counting how often each word appears in the document, giving a structured data format that can be used in machine learning models.

The return type of CountVectorizer is typically a sparse matrix of type "scipy.sparse.csr_matrix" by default. This matrix contains the frequency counts of each word in the document collection, with rows representing documents and columns representing words.



```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split(' '):
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# Now the problem with vectorization is that simiar meaning english words still gets separated
# for example, actor and actors means same thing but will treated as different words, similarly, actual&actually etc...
# To tackle this problem I will be using stemming
# What stemming does is that it will change similar words to a sing word 
# for example; dance, dancer, and dancing ----> dance, dance, dance
```

```python
# So with the vectors we will have 4086 movies --> 4086 vectors in 5000 dimensional space
# So in higher dimension, ecuclidean distance fails
# Euclidean distance becomes less reliable in higher dimensions because of the curse of dimensionality. 
# In high-dimensional spaces, points tend to become equidistant, making it hard to distinguish between them based on distance alone. 
# This results in the distance losing its discriminative power as a measure of similarity.
# The curse of dimensionality refers to how adding more dimensions (features) to data can make it harder to analyze because the data becomes 
# sparse and distances between points become less meaningful.

# Hence we will use COSINE distance i.e. angle between two vectors. the less the angle the more the similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x : x[1])[1:11]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x : x[1])[1:11]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
import requests
```

```python
def fetch_poster(movie_id):
    url = "https://api.themoviedb.org/3/movie/{}?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US".format(movie_id)
    data = requests.get(url)
    data = data.json()
    poster_path = data['poster_path']
    full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
    return full_path
```

```python
def recommend(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    recommended_movie_names = []
    recommended_movie_posters = []
    for i in distances[1:6]:
        # fetch the movie poster
        movie_id = movies.iloc[i[0]].movie_id
        recommended_movie_posters.append(fetch_poster(movie_id))
        recommended_movie_names.append(movies.iloc[i[0]].title)

    return [recommended_movie_names,recommended_movie_posters]

recommended_movie_list = recommend('Batman Begins')

for i in range(len(recommended_movie_list[0])):
    print(recommended_movie_list[0][i])
    print(recommended_movie_list[1][i])
```
